```python
#drop column
df_pyspark.drop('Other technologies/programming languages you use often').show()
```

```python
#drop null values 
df_pyspark.na.drop().show()
```

```python
### any = how
df_pyspark.na.drop(how='all').how()
```

```python
#threshold
df_pyspark.na.drop(how='any').how()
```

```python
#threshold limit 
df_pyspark.na.drop(how='any',thresh=1).how()
```

```python
df_pyspark.na.drop(how='all', thresh=5, subset=['Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week']).how()
```

```python
#fill missing value
df_pyspark.na.fill('Missing Value').show()
```

```python
#fill missing value with mean,median,mode 

from pyspark.ml.feature import Imputer

Imputer = Imputer(
    inputCols=['Age'],
    outputCols=['{}_imputed'.format(c) for c in ['Age']]
).setStrategy('mean')
```

```python
#add imputed columns to df
Imputer.fit(df_pyspark).transform(df_pyspark).show()
```

```python
#fill missing value with mean

from pyspark.ml.feature import Imputer

Imputer = Imputer(
    inputCols=['Years of experience in Germany'],
    outputCols=['{}_imputed'.format(c) for c in ['Years of experience in Germany']]
).setStrategy('mean')
```

```python
#add imputed columns to df
Imputer.fit(df_pyspark).transform(df_pyspark).show()
```
